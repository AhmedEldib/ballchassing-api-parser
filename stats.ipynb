```python
import json
import glob
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
matches = []

for j in glob.glob('jsons/*'):
    with open(j) as f:
        matches.append(json.load(f))

team = matches[0]['matches'][0]['blue']['players'][0]
team
```

```python
player_data = []

for day in matches:
    for match in day['matches']:
        for team in ['blue', 'orange']:
            for player in match[team]['players']:
                parsed_dict = {}

                parsed_dict['name'] = player['name']
                parsed_dict['platform'] = player['id']['platform']
                parsed_dict['id'] = player['id']['id']
                parsed_dict['team_name'] = match[team]['name']
                parsed_dict['car_name'] = player['car_name']

                parsed_dict['avg_speed'] = player['stats']['movement']['avg_speed']
                parsed_dict['total_distance'] = player['stats']['movement']['total_distance']
                
                parsed_dict['shots'] = player['stats']['core']['shots']
                parsed_dict['shots_against'] = player['stats']['core']['shots_against']
                parsed_dict['goals'] = player['stats']['core']['goals']
                parsed_dict['goals_against'] = player['stats']['core']['goals_against']
                parsed_dict['saves'] = player['stats']['core']['saves']
                parsed_dict['assists'] = player['stats']['core']['assists']
                parsed_dict['score'] = player['stats']['core']['score']
                parsed_dict['mvp'] = player['stats']['core']['mvp']
                parsed_dict['shooting_percentage'] = player['stats']['core']['shooting_percentage']

                parsed_dict['demos_inflicted'] = player['stats']['demo']['inflicted']
                parsed_dict['demos_taken'] = player['stats']['demo']['taken']

                player_data.append(parsed_dict)
```

```python
data = pd.DataFrame(data=player_data, columns=player_data[0].keys())
data.head(5)
```

```python
data.to_csv('stats.csv')
```
